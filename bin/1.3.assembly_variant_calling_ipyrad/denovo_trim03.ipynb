```python
#Importar librerias
import ipyrad as ip
import ipyparallel as ipp
```

```python
## Crear un objeto Ensamble llamado data1. 
denovo_trim03 = ip.Assembly("denovo_trim03")
```

```python
## prints the parameters to the screen
denovo_trim03.get_params()
```

```python
## setting/modifying parameters for this Assembly object
denovo_trim03.set_params('project_dir', '../data/1.3.assemble_variant_calling/1.3.2.ipyrad/denovo_trim03')
denovo_trim03.set_params('sorted_fastq_path', "../data/1.1.filter/*trim03.fastq.gz")
denovo_trim03.set_params('assembly_method', 'denovo')
denovo_trim03.set_params('datatype', 'ddrad')
denovo_trim03.set_params('output_formats', ['p', 's', 'k', 'g', 'v'])

## prints the parameters to the screen
denovo_trim03.get_params()
```

```python
## run step 1 to create Samples objects
denovo_trim03.run("1", auto=True, force=True)
```
