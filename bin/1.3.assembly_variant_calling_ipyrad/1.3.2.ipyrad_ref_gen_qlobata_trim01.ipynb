```python
#Importar librerias
import ipyrad as ip
import ipyparallel as ipp
```

```python
## Crear un objeto Ensamble llamado data1. 
ref_gen_qlob_trim01 = ip.Assembly("ref_gen_qlob_trim01")
```

```python
## prints the parameters to the screen
ref_gen_qlob_trim01.get_params()
```

```python
## setting/modifying parameters for this Assembly object
ref_gen_qlob_trim01.set_params('project_dir', '../data/1.3.assemble_variant_calling/1.3.2.ipyrad/ref_gen_qlob_trim01')
ref_gen_qlob_trim01.set_params('sorted_fastq_path', "../data/1.1.filter/*trim01.fastq.gz")
ref_gen_qlob_trim01.set_params('assembly_method', "reference")
ref_gen_qlob_trim01.set_params('reference_sequence', '../data/reference_genomes/Qlobata.v3.0.RptMsk4.0.6.on-RptMdl1.0.8.softmasked.fasta')
ref_gen_qlob_trim01.set_params('datatype', 'ddrad')
ref_gen_qlob_trim01.set_params('output_formats', ['p', 's', 'k', 'g', 'v'])

## prints the parameters to the screen
ref_gen_qlob_trim01.get_params()
```

```python
## run step 1 to create Samples objects
ref_gen_qlob_trim01.run("1", auto=True, force=True)
```
